```python
import os
# Create a directory to store the text files
output_dir = "customtextclassification"

# Save each category and its entries to separate text files
for category, entries in categories.items():
    for i, entry in enumerate(entries, start=1):
        filename = f"{category.replace(' ', '_')}_{i}.txt"
        filepath = os.path.join(output_dir, filename)
        with open(filepath, "w") as file:
            file.write(entry)

print(f"Text files saved in the '{output_dir}' directory.")
```

```python
from azure.storage.blob import ContainerClient

# SAS URL for the container
sas_url = "https://ziggylanguagestorage.blob.core.windows.net/emails?sp=rl&st=2025-04-15T18:12:02Z&se=2025-04-18T02:12:02Z&spr=https&sv=2024-11-04&sr=c&sig=JL9mynxJp8goydwPFNgANoO3LnTvlkyKFRQrQqRQaQs%3D"  # Replace with your SAS URL for the container

# Initialize the ContainerClient using the SAS URL
container_client = ContainerClient.from_container_url(sas_url)

# Count the number of blobs in the container
blob_count = 0
for blob in container_client.list_blobs():
    blob_count += 1

print(f"Number of files in the container: {blob_count}")
```
